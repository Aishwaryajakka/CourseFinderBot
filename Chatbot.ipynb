```python
from nltk.chat.util import Chat, reflections
```

```python
reflections = {
  "i am"       : "you are",
  "i was"      : "you were",
  "i"          : "you",
  "i'm"        : "you are",
  "i'd"        : "you would",
  "i've"       : "you have",
  "i'll"       : "you will",
  "my"         : "your",
  "you are"    : "I am",
  "you were"   : "I was",
  "you've"     : "I have",
  "you'll"     : "I will",
  "your"       : "my",
  "yours"      : "mine",
  "you"        : "me",
  "me"         : "you"
}
```

```python
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, Good Morning!",]
    ],
     [
        r"what is your name ?",
        ["My name is Roc and I'm a Pitt chatbot.",]
    ],
    [
        r"how are you ?",
        ["I'm doing good\nHow about You ?",]
    ],
    [
        r"I am good.",
        ["Nice to hear that.",]
    ],
    [
        r"sorry (.*)",
        ["Its alright","Its OK, never mind",]
    ],
    [
        r"i'm (.*) doing good",
        ["Nice to hear that","Alright :)",]
    ],
    [
        r"hi|hey|hello",
        ["Hello", "Hey there",]
    ],
    [
        r"(.*) age?",
        ["I'm a computer program dude\nSeriously you are asking me this?",]
        
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse",]
        
    ],
    [
        r"(.*) created ?",
        ["Nagesh created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",
        ['Pittsburgh, Pennsylvania',]
    ],
    [
        r"how is weather in (.*)?",
        ["Weather in %1 is awesome like always","Too hot man here in %1","Too cold man here in %1","Never even heard about %1"]
    ],
    [
        r"i work in (.*)?",
        ["%1 is an Amazing company, I have heard about it. But they are in huge loss these days.",]
    ],
[
        r"(.*)raining in (.*)",
        ["No rain since last week here in %2","Damn its raining too much here in %2"]
    ],
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy ",]
    ],
    [
        r"(.*) (sports|game) ?",
        ["I'm a very big fan of Football",]
    ],
    [
        r"who (.*) sportsperson ?",
        ["Messy","Ronaldo","Roony"]
],
    [
        r"who (.*) (moviestar|actor)?",
        ["Brad Pitt"]
],
    
    [
        r"Tell me something about the course 2711?",
        ["This is a graduate level Advanced Database course (3 credits), usually taught in Spring semesters. "]
],
    [
        r"Is there any pre-requisite for the course 2711?",
        ["Yes, you either need to take INFSCI 2710 (Database Management Systems) or you need to have a fair amount of knowledge of relational databases. "]
],
    [
        r"What is the difficulty level of course 2711?",
        ["This course should not be difficult if you are strong enough in relational databases."]
],
    [
        r"How is the course 2711 going to be helpful?",
        ["Yes, if you are planning to explore in Big Data field or Data Warehousing, then this is a must-take course."]
],
     [
        r"Is there any other course recommended along with 2711?",
        ["Yes, Data Mining makes a good match with Advanced Database."]
],
     [
        r"Interesting, why do you recommend Data Mining?",
        ["Well, Data Mining is very helpful in implementing data pre-processing techniques used for Machine Learning Algorithms. All the data scientists are supposed to have a good knowledge of Data Mining before they start working."]
],
    [
        r"What knowledge do you think is important for enrolling in Data Mining course?",
        ["A fair knowledge of Data Analytics and Math foundation is important to have. Math part would include statistics, calculus and linear algebra."]
],
    [
        r"Tell me few more courses I should enroll into if I want to take Information Science Major along with Big Data Analytics concentration?",
        ["Data Analytics, Data Mining, Algorithm Design and Relational Databases are mandatory courses for Big Data Analytics concentration. Other recommended courses include Cryptography, Advanced Databases, Information Visualization and Cloud Computing."]
],
    [
        r"quit",
        ["BBye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
],
]


def roc():
    print("Hi, I'm Roc and I am here to provide you information about various courses lot about University of Pittsburgh ;)\nPlease type lowercase English language to start a conversation. Type quit to leave ") #default message at the start
    chat = Chat(pairs, reflections)
    chat.converse()
    if __name__ == "__main__":
        roc()
```
